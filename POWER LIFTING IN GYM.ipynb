```python
data.columns =data.columns.str.strip()
lifttype1 = data.loc[data['Lift Type'] == "Bench Press"].count()[0]
lifttype2 = data.loc[data['Lift Type'] == "Squat"].count()[0]
lifttype3 = data.loc[data['Lift Type'] == "Deadlift"].count()[0]
colors = ['blue', 'orange','green']  
plt.pie([lifttype1, lifttype2,lifttype3], labels=['Bench Press', 'Squat','Deadlift'], colors=colors, autopct='%.2f%%')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount Lifted (kg)'], bins=20, kde=True)
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount Lifted (kg)'], bins=20, kde=True)

plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.scatterplot(data, x='Age', y='Amount Lifted (kg)')
plt.xlabel('Age')
plt.ylabel('Amount Lifted (kg)')
```

```python
sns.boxplot(data=data, x='Weight Class', y='Lifter Name')
plt.xlabel('Weight Class')
plt.ylabel('Lifter Name')
```

```python
weight = data.groupby('Age')['Amount Lifted (kg)'].mean()

plt.figure(figsize=(12,8))
weight.plot(kind = 'line',marker = 'D',color = 'green', mec = 'r', ms = 10)
plt.xlabel('Age')
plt.ylabel('Average Amount Lifted (kg)')
plt.title('Average Weight Lifted by Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
avg_lift_by_type = data.groupby('Lift Type')['Amount Lifted (kg)'].mean()
avg_lift_by_type.plot(kind='bar')
plt.xlabel('Lift Type')
plt.ylabel('Average Amount Lifted (kg)')
plt.title('Average Amount Lifted by Lift Type')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
data.boxplot(column='Amount Lifted (kg)', by='Weight Class', grid=False)
plt.xlabel('Weight Class')
plt.ylabel('Amount Lifted (kg)')
plt.title('Box Plot of Amount Lifted by Weight Class')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
lift_type_counts = data['Lift Type'].value_counts()
plt.pie(lift_type_counts, labels=lift_type_counts.index, autopct='%1.1f%%', colors=plt.get_cmap('tab20').colors)
plt.title('Distribution of Lift Types')
plt.show()
```

```python
sns.histplot(data, x='Amount Lifted (kg)', hue='Lift Type', multiple='stack', palette='viridis')
plt.title('Distribution of Amount Lifted by Lift Type')
plt.xlabel('Amount Lifted (kg)')
plt.ylabel('Count')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(data[['Age', 'Amount Lifted (kg)', 'Weight Class']], hue='Weight Class')
plt.suptitle('Pair Plot of Age, Amount Lifted, and Weight Class', y=1.02)
plt.show()
```
